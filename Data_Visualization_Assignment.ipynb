5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns


# Generate synthetic data with correlated features
np.random.seed(0)  # For reproducibility

# Number of samples
n_samples = 100

# Generate features with correlation
# Feature 1 (x1) is random
x1 = np.random.normal(loc=10, scale=5, size=n_samples)

# Feature 2 (x2) is correlated with x1 (e.g., x2 = x1 + noise)
x2 = x1 + np.random.normal(loc=0, scale=2, size=n_samples)

# Feature 3 (x3) is somewhat correlated with x1 but less so
x3 = 0.5 * x1 + np.random.normal(loc=0, scale=5, size=n_samples)

# Feature 4 (x4) is random and not correlated with the others
x4 = np.random.normal(loc=5, scale=3, size=n_samples)

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd

# Generate x and y data using np.linspace
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid for X and Y
x, y = np.meshgrid(x, y)

# Compute z based on the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data to create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

# Create the 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis',  # Set a color scale
    colorbar=dict(title='Z Value')  # Label for the colorbar
)])

# Update plot layout
fig.update_layout(
    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    autosize=True
)

# Show the plot
fig.show()
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.


```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data for x and y, as well as sizes and colors for the markers
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 30, 100)
colors = np.random.choice(['blue', 'green', 'red', 'purple'], 100)

# Create a data source
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Random Scatter Plot", width=700, height=400)
p.scatter('x', 'y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

```python
# Updated data for the bar chart
from bokeh.models import FactorRange # Import FactorRange from bokeh.models
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=FactorRange(*fruits), title="Fruit Counts", width=700, height=400)
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="coral")

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.layouts import layout
import numpy as np

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh's ColumnDataSource
heatmap_data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'values': data_heatmap.flatten()
}

# Color mapper for the heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=heatmap_data['values'].min(), high=heatmap_data['values'].max())

# Create the heatmap plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)
p.rect(x='x', y='y', width=1/10, height=1/10, source=heatmap_data, fill_color=transform('values', color_mapper), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title="Values")
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
